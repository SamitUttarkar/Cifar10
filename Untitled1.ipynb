```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Churn_Modelling.csv')
```

```python
df
```

```python
X = df.iloc[:,3:13]
y = df.iloc[:,13]
```

```python
geography = pd.get_dummies(X['Geography'], drop_first = True)
gender = pd.get_dummies(X['Gender'],drop_first = True)
```

```python
X.drop(['Geography','Gender'],axis = 1, inplace = True)
```

```python
X = pd.concat([X,geography,gender],axis = 1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# ANN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU, ReLU
from tensorflow.keras.layers import Dropout
```

```python
# Sequential - we will be able to do forward propogation and  backward propogation 
# dense - we will be able to create the layers - input and hidden layers 
# ReLU - activation function 
# dropout layer - Sometimes the model leads to overfitting - it is a regulatization parameter - eg dropu layer = 0.3 means 30% of the neuron in the layer will be deactivated
```

```python
# Initialize the ANN
classifier = Sequential()
```

```python
## adding input layers 
classifier.add(Dense(units = 11,activation = 'relu')) # units = 11 means number of columns 
# relu will be automatically applied to the next layers }
#classifier.add(Dropout(0.2))
```

```python
# adding the first hidden layer
classifier.add(Dense(units = 7,activation = 'relu'))
```

```python
classifier.add(Dense(units = 6,activation = 'relu'))
```

```python
# Adding output layer
classifier.add(Dense(units = 1 ,activation = 'sigmoid')) # binary classification therefore 1 neuron output
# this is a binary classificaion therefore activation sigmoid instead of softmax
```

```python
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['Accuracy'])
# bydefault adam uses learning rate of 0.1l
```

```python
import tensorflow as tf
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.0001,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```

```python
model_history = classifier.fit(X_train,y_train,validation_split = 0.33, batch_size = 10, epochs = 100, callbacks = early_stopping)
```

```python
# we will use early stopping to decide the number of epoches l
model_history.history.keys()
```

```python
plt.plot(model_history.history['Accuracy'])
plt.plot(model_history.history['val_Accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.ylim(0.80, 0.88)
plt.legend(['train','test'], loc = 'upper left')
plt.show()
```

```python
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
#plt.ylim(0.80, 0.88)
plt.legend(['train','test'], loc = 'upper left')
plt.show()
```

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred >= 0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm =  confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_pred,y_test)
```
